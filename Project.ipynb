```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
%matplotlib inline

# StatsModels
import statsmodels.api as sm
import statsmodels.formula.api as smf

# sklearnModels
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm
from sklearn.svm import SVR
from sklearn.datasets import make_regression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import RFE
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
from sklearn.datasets import load_boston
```

```python
data = pd.read_csv('Data_Cleaning.csv')
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
np.unique(df['Type_of_snake'])
```

```python
df1 = df[df['Type_of_snake']=='corn_king_milk']
y = df['Egg_age'].to_numpy()
X = df.drop(['Egg_age', 'mean_pulse_rate', 'Eggscore', 'Type_of_snake'], axis=1).to_numpy()
```

```python
scores = []
best_mlr = LinearRegression()
cv = KFold(n_splits=10, random_state=50, shuffle=False)
for train_index, test_index in cv.split(X,y):
    print("Train Index: ", train_index, "\n")
    print("Test Index: ", test_index)

    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    best_mlr.fit(X_train, y_train)
    scores.append(best_mlr.score(X_test, y_test))
```

```python
print("LR coefficient is", best_mlr.coef_)
print("LR intercept is", best_mlr.intercept_)
```

```python
print(np.mean(scores))
```

```python
cross_val_score(best_mlr, X, y, cv=10)
```

```python
cross_val_predict(best_mlr, X, y, cv=10)
```

```python
y_pred =  best_mlr.predict(X_test)
y_pred
```

```python
# importing r2_score module
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
# predicting the accuracy score
print("r2 score is", r2_score(y_test,y_pred))
print("MAE is", mean_absolute_error(y_test, y_pred))
print("MSE is", mean_squared_error(y_test,y_pred))
print("RMSE is", np.sqrt(mean_squared_error(y_test,y_pred)))
```
